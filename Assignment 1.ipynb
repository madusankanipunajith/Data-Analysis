```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import linear_kernel
```

```python
# opening the text files and copying to variables
def make_string_from_file(file_name):
    text = ''
    with open(file_name,"r") as provided_file:
        for line in provided_file:
   
            # reading each word        
            for word in line.split():
                text = text + word + ' '  
        
        return text.strip()
    
def get_unique_words_from_file(file_name):
        text_1 = make_string_from_file(file_name)
        unique_words = set(text_1.split(' '))
        str_val = " ".join(unique_words)
        return str_val
```

```python
doc_1 = make_string_from_file("Downloads/doc 1.txt")
doc_2 = make_string_from_file("Downloads/doc 2.txt")
doc_3 = make_string_from_file("Downloads/doc 3.txt")
doc_4 = make_string_from_file("Downloads/doc 4.txt")
doc_5 = make_string_from_file("Downloads/doc 5.txt")
doc_6 = make_string_from_file("Downloads/doc 6.txt")
doc_7 = make_string_from_file("Downloads/doc 7.txt")
doc_8 = make_string_from_file("Downloads/doc 8.txt")

print(doc_8)
```

```python
corpus = [doc_1, doc_2, doc_3, doc_4, doc_5, doc_6, doc_7, doc_8]
```

```python
# TF-IDF genaration
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(corpus)
feature_names = vectorizer.get_feature_names_out()
print(feature_names, len(feature_names))
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
```

```python
# check cosine similarities of document 1 with all 8 documents
cosine_similarities = linear_kernel(document_1, vectors).flatten()
cosine_similarities
```

```python
# define a function to print all similarities at once
def get_cos_sim(doc, number):
    print(f'Document : {number}.txt')
    cosine_similarities = linear_kernel(doc, vectors).flatten()
    print(cosine_similarities, '\n')


def print_similarities(number):
    match number:
        case 1: 
            return get_cos_sim(document_1, 1)
        case 2: 
            return get_cos_sim(document_2, 2)
        case 3: 
            return get_cos_sim(document_3, 3)
        case 4: 
            return get_cos_sim(document_4, 4)
        case 5: 
            return get_cos_sim(document_5, 5)
        case 6: 
            return get_cos_sim(document_6, 6)
        case 7: 
            return get_cos_sim(document_7, 7)
        case 8: 
            return get_cos_sim(document_8, 8)
        case default:
            print('something went wrong')
            return

        
```

```python
for doc_num in range (1,9):
    print_similarities(doc_num)
```
