```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import linear_kernel
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
cachedStopWords = stopwords.words("english")
```

```python
# opening the text files and copying to variables
def make_string_from_file(file_name):
    text = ''
    with open(file_name,"r") as provided_file:
        for line in provided_file:
   
            # reading each word        
            for word in line.split():
                text = text + word + ' '  
        
        return text.strip()
    
def get_unique_words_from_file(file_name):
        text_1 = make_string_from_file(file_name)
        unique_words = set(text_1.split(' '))
        str_val = " ".join(unique_words)
        return str_val
```

```python
doc_1 = make_string_from_file("Downloads/doc 1.txt")
doc_2 = make_string_from_file("Downloads/doc 2.txt")
doc_3 = make_string_from_file("Downloads/doc 3.txt")
doc_4 = make_string_from_file("Downloads/doc 4.txt")
doc_5 = make_string_from_file("Downloads/doc 5.txt")
doc_6 = make_string_from_file("Downloads/doc 6.txt")
doc_7 = make_string_from_file("Downloads/doc 7.txt")
doc_8 = make_string_from_file("Downloads/doc 8.txt")

print(doc_8)
```

```python
corpus = [doc_1, doc_2, doc_3, doc_4, doc_5, doc_6, doc_7, doc_8]
```

```python
# TF-IDF genaration
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(corpus)
feature_names = vectorizer.get_feature_names_out()
print(feature_names, len(feature_names))
for text in feature_names:
    print(text)
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
```

```python
# custom function for identify valid topics
# return a list
def valid_topic(topic):
    valid_topic = []
    for text in topic.split():
        try:
            output = df[text.lower()]
            valid_topic.append(text.lower())
        except:
            pass
    return valid_topic
```

```python
# define a function to print all similarities at once
def get_cos_sim(doc, number):
    print(f'Document : {number}.txt')
    cosine_similarities = linear_kernel(doc, vectors).flatten()
    print(cosine_similarities, '\n')


def print_similarities(number):
    match number:
        case 1: 
            return get_cos_sim(document_1, 1)
        case 2: 
            return get_cos_sim(document_2, 2)
        case 3: 
            return get_cos_sim(document_3, 3)
        case 4: 
            return get_cos_sim(document_4, 4)
        case 5: 
            return get_cos_sim(document_5, 5)
        case 6: 
            return get_cos_sim(document_6, 6)
        case 7: 
            return get_cos_sim(document_7, 7)
        case 8: 
            return get_cos_sim(document_8, 8)
        case default:
            print('something went wrong')
            return

        
```

```python
for doc_num in range (1,9):
    print_similarities(doc_num)
```

```python
def remove_stop_words(doc):
  
    stop_words = set(stopwords.words('english'))
  
    word_tokens = word_tokenize(doc)
  
    tokens_without_sw = []
  
    for w in word_tokens:
        if w not in stop_words:
            tokens_without_sw.append(w)
    
    filtered_sentence = (" ").join(tokens_without_sw)
    return filtered_sentence
```

```python
new_doc_1 = remove_stop_words(doc_1)
new_doc_2 = remove_stop_words(doc_2)
new_doc_3 = remove_stop_words(doc_3)
new_doc_4 = remove_stop_words(doc_4)
new_doc_5 = remove_stop_words(doc_5)
new_doc_6 = remove_stop_words(doc_6)
new_doc_7 = remove_stop_words(doc_7)
new_doc_8 = remove_stop_words(doc_8)

new_corpus = [new_doc_1, new_doc_2, new_doc_3, new_doc_4, new_doc_5, new_doc_6, new_doc_7, new_doc_8]
```

```python
new_vectors = vectorizer.fit_transform(new_corpus)
new_feature_names = vectorizer.get_feature_names_out()
new_dense = new_vectors.todense()
new_denselist = new_dense.tolist()
new_df = pd.DataFrame(new_denselist, columns=new_feature_names)
```

```python
# get cosine similarity matrix by removing stop words in the docs
print(cosine_similarity(new_df.values, new_df.values))
```
